# IMPORTS

```python
# general
from pprint import pprint
from pathlib import Path
import random
from IPython.display import Image, display

# main
from parseimagenet import get_image_paths_by_keywords
from parseimagenet import get_available_presets, KEYWORD_PRESETS

# check
!pip show parseimagenet
```

# USE

```python
# Set the path to your ImageNet directory
base_path = Path('/Users/mrt/Documents/MrT/code/computer-vision/image-bank/ImageNet-Subset')
```

## Explore

```python
# See list of available preset names
print(get_available_presets())
```

```python
# Inspect the actual keywords in a preset (first 10)
print(f'Bird Keywords:\n{KEYWORD_PRESETS["birds"][:10]}\n')
print(f'Snake Keywords:\n{KEYWORD_PRESETS["snakes"][:10]}\n')
print(f'Dog Keywords:\n{KEYWORD_PRESETS["dogs"][:10]}\n')
print(f'Wild Canid Keywords:\n{KEYWORD_PRESETS["wild_canids"][:10]}\n')
```

## Fetch Random Images

```python
train_random_image_paths = get_image_paths_by_keywords(
    base_path=base_path,
)
pprint(train_random_image_paths[:2])
```

## Fetch `Train` Images

```python
train_dog_image_paths = get_image_paths_by_keywords(
    base_path=base_path,
    source='train',
    preset="dogs",
    num_images=100,
    silent=False,
)
pprint(train_dog_image_paths[:10])
```

```python
train_dog_random_index = random.randrange(len(train_dog_image_paths))
display(Image(filename=train_dog_image_paths[train_dog_random_index]))
```

## Fetch `Val` Images

```python
val_bird_image_paths = get_image_paths_by_keywords(
    base_path=base_path,
    source='val',
    keywords=["American eagle"],
    num_images=100,
)
pprint(val_bird_image_paths[:10])
```

```python
val_bird_random_index = random.randrange(len(val_bird_image_paths))
display(Image(filename=val_bird_image_paths[val_bird_random_index]))
```
