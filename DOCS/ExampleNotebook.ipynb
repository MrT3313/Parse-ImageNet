# USE

```python
# Set the path to your ImageNet directory
base_path = Path('/Users/mrt/Documents/MrT/code/computer-vision/image-bank/ImageNet-Subset')
```

## Fetch `Train` Images

```python
train_dog_image_paths = get_image_paths_by_keywords(
    base_path=base_path,
    source='train', # or 'val' | the Kaggle Dataset does not provide the ground truth for the testing data
    preset="dogs",
    num_images=100,
    silent=False,
)
pprint(train_dog_image_paths[:10])
```

```python
train_dog_random_index = random.randrange(len(train_dog_image_paths))
display(Image(filename=train_dog_image_paths[train_dog_random_index]))
```

## Fetch `Val` Images

```python
val_bird_image_paths = get_image_paths_by_keywords(
    base_path=base_path,
    source='val', # or 'test' | the Kaggle Dataset does not provide the ground truth for the testing data
    preset="birds",
    num_images=100,
)
pprint(val_bird_image_paths[:10])
```

```python
val_bird_random_index = random.randrange(len(val_bird_image_paths))
display(Image(filename=val_bird_image_paths[val_bird_random_index]))
```
